```python
# T Test
# H0=There is no difference in sales after digital marketing
#H1= There is a difference in sales after digital marketing
```

```python
## Hypothesis testing
ttest,pval = stats.ttest_rel(sales_add['Sales_before_digital_add(in $)'], sales_add['Sales_After_digital_add(in $)'])
print(pval)
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
# From the T Test there is an increase in sales after digital marketing
```
